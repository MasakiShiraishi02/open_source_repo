Quantum annealing will find the minimum of the following energy function and the best 'x' combination that gives the minimum energy at that time.
$$
E({\bf x}) = \sum_{i=1}^N \sum_{j=1}^N Q_{ij}x_ix_j
$$

These are spin glass models. The folowing are examples of calculation using quantum annealing.

```python
import numpy as np
```

```python
N = 50
QUBO = np.random.randn(N**2).reshape(N, N)
QUBO = (QUBO + QUBO.T) / 2
```

```python
print(QUBO)
```

```python
from openjij import SQASampler
```

```python
sampler = SQASampler()
```

```python
sampleset = sampler.sample_qubo(QUBO, num_sweeps=10000, num_reads=100)
```

```python
print(sampleset.record)
```

```python
energy = sampleset.data_vectors['energy']
print(energy)
```

```python
import matplotlib.pyplot as plt
plt.hist(energy, bins=100)
plt.show()
```

```python
from openjij import SASampler
sampler = SASampler()
sampleset = sampler.sample_qubo(QUBO, num_sweeps=10000, num_reads=100)
energy = sampleset.data_vectors['energy']
plt.hist(energy, bins=100)
plt.show()
```

In the following, the QUBO matirix is built using the ising model as an example.

```python
N = 500
Jmat = np.zeros((N, N))
for i in range(N):
    for j in range(N):
        Jmat[i, j] = -1/(2*N)
```

```python
hvec = np.zeros(N)
```

```python
import dimod
model = dimod.BinaryQuadraticModel(hvec, Jmat, 0.0,  'SPIN')
qubo, offset = model.to_qubo()
```

```python
qubo
```

```python
sampler = SASampler()
sampleset = sampler.sample_qubo(qubo, beta_max=0.1, num_sweeps=10000, num_reads=1)
```

```python
sampleset.record[0][0]
```

```python
beta_list = np.linspace(0.1,2.0,20)
```

```python
beta_list
```

```python
num_reads = 1000
mag_list = []
var_list = []

for beta in beta_list:
   sampleset = sampler.sample_qubo(qubo, beta_max=beta, num_sweeps=10000, num_reads=num_reads)
   m2_list = []
   for k in range(num_reads):
        binary = sampleset.record[k][0]
        spin = 2*binary - 1
        m2 = spin.mean()**2
        m2_list.append(m2)
   m2_array = np.array(m2_list)
   print(beta, m2_array.mean(), m2_array.var())
   mag_list.append(m2_array.mean())
   var_list.append(m2_array.var())    
```

```python
plt.plot(beta_list, mag_list)
plt.show()
```
